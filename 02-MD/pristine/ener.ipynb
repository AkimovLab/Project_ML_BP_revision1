# Running molecular dynamics with CP2K

In this tutorial for molecular dynamics calculations using DFT. The rutile $TiO_2$ unit cell is considered. VESTA is used to build the molecular structure model. We first downloaded its `.cif` file from [this link](https://materialsproject.org/materials/mp-2657/) and then loaded and saved its coordinates in VASP POSCAR format in Cartesian coordinates. In order to use this file in CP2K, we will need to generate an `.xyz` file from this file and also insert the correct cell size in the input file. The `libra_py.data_conv.vasp_to_xyz` function will do it for us and prints out the cell A, B, and C vectors.

## 4. Run CP2K <a name="run_cp2k"></a>
[Back to TOC](#toc)

To run CP2K, you need to load all the necessary libraries used to compile it. More information on how to run CP2K and the inputs can be found in [here](https://github.com/compchem-cybertraining/Tutorials_CP2K).

If you want to submit the calculations on a node, you will need to submit your job using the provided slurm file. This is done by `sbatch submit.slm`. If you want to run the calculations on your local terminal what you need to do is to run the following commands.

$\textbf{Note:}$ This step might take some time to complete. We set the calculations for MD with 50 steps here. To reduce the number of steps you can change the `&MOTION->&MD->STEPS` for shorter MD in the input file. Also, if you use a cluster and you are running the Jupyter notebook, your calculations may be terminated in a couple of minutes. So, it is always better to submit a job if you access to a cluster by `sbatch submit.slm`. For example, as you can see below, we use 16 number of processors and it takes about 8 seconds per step but the job terminates automatically after a couple of minutes by the cluster (by `kill -9 processID`). The timing data is shown in the `*.ener` file.

```python
! module load intel/20.2
! module load intel-mpi/2020.2
! mpirun -np 16 /projects/academic/cyberwksp21/Software/cp2k-intel/cp2k-8.2/exe/Linux-x86-64-intelx/cp2k.psmp -i md.inp -o out_md_tio2_dft.log
! echo "Done!"
```

$\textbf{Note:}$ It might happen that you get the error `mpirun not found`. If this happened, exit Jupyter notebook and load the libraries then rerun the Jupyter and the above cell.

## 5. Plot properties <a name="plot_properties"></a>
[Back to TOC](#toc)

You can start plotting some time-dependent results including the temperature vs time. This is done by using the `.ener` file.
You can also plot some other time-depedent properties as well to properly check the equilibrium of your system including total energy vs time etc. The current system is clearly not equilibrated up to 50 steps.

```python
%matplotlib notebook
data = np.loadtxt('MD_p4-1.ener')
plt.plot(data[:,1],data[:,3])
plt.xlabel('Time, fs')
plt.ylabel('Temperature, K')
plt.title('  Pristine ML_BP')
plt.tight_layout()
plt.savefig('Pristine_ML_BP_MD_Temperature_300K', dpi=600)
```
